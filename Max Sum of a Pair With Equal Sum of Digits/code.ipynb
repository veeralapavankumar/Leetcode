```python
class Solution(object):
    def maximumSum(self, nums):
        digit_sum_map = {}

        for num in nums:
            s = 0
            temp = num
            while temp > 0:
                s += temp % 10
                temp //= 10

            if s not in digit_sum_map:
                digit_sum_map[s] = []
            digit_sum_map[s].append(num)

        max_sum = -1
        for values in digit_sum_map.values():
            if len(values) >= 2:
                values.sort(reverse=True)
                max_sum = max(max_sum, values[0] + values[1])

        return max_sum
```
