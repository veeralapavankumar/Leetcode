```python
class Solution(object):
    def isMonotonic(self, nums):
        NonDecreasing = True
        NonIncreasing = True
        for i in range(1,len(nums)):
            if nums[i] < nums[i-1]:
                NonDecreasing = False
            if nums[i] > nums[i-1]:
                NonIncreasing = False
        return NonDecreasing or NonIncreasing
        
        
```
